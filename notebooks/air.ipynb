```python
import pandas as pd
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
from sklearn.impute import KNNImputer

# Load the data
df = pd.read_csv("Raw_data_1Hr_2023_site_5345_Sector-51_Gurugram_HSPCB_1Hr (1).csv")

# Parse the Timestamp
df['Timestamp'] = pd.to_datetime(df['Timestamp'], format='mixed')
df.set_index('Timestamp', inplace=True)
# Drop columns with all NaN values to simplify the imputation process
df = df.dropna(axis=1, how='all')
# Separate the index and the data
df_index = df.index
numerical_data = df.values
df
```

```python


# Impute missing values using KNNImputer
imputer = KNNImputer(n_neighbors=5)
imputed_data = imputer.fit_transform(numerical_data)

# Create a DataFrame with the imputed data and original column names and index
df_imputed = pd.DataFrame(imputed_data, columns=df.columns, index=df_index)

# Verify the Data
df_imputed
```

```python
from statsmodels.tsa.api import VAR
from sklearn.preprocessing import StandardScaler
from statsmodels.tsa.stattools import adfuller

data_diff = df_imputed.diff().dropna()
# Scale the data
scaler = StandardScaler()
data_scaled = pd.DataFrame(scaler.fit_transform(data_diff), columns=data_diff.columns)

# Check for stationarity
for column in data_scaled.columns:
    result = adfuller(data_scaled[column])
    print(f'{column}: p-value = {result[1]}')

# Fit VAR model
model = VAR(data_scaled)
lag_order = model.select_order(maxlags=15)
print(lag_order.summary())

# Select the best lag order
model_fitted = model.fit(lag_order.aic)
print(model_fitted.summary())

# Check residuals
residuals = model_fitted.resid
print(residuals.describe())

# Diagnose residuals
from statsmodels.stats.diagnostic import acorr_ljungbox
ljung_box_results = acorr_ljungbox(residuals, lags=[10])
print(ljung_box_results)

# Impulse Response Analysis
irf = model_fitted.irf(10)
irf.plot(orth=False)

# Forecasting
forecast = model_fitted.forecast(data_scaled.values[-lag_order.aic:], steps=10)
forecast = scaler.inverse_transform(forecast)
forecast_df = pd.DataFrame(forecast, columns=data_diff.columns)
print(forecast_df)
```
